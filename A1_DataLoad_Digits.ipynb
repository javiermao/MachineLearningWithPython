2. The dataset description
    - This dataset is made up of 1797 8x8 images
    - Each image is of a hand-written digit (0,1,...9)
    - When uploading the dataset, each image is a 8x8 matrix (or 64 pixels)
    - In some situations, we can transform the 8x8 matrix into a a feature vector of length 64


# Importing and inspecting the data

```python
# Import the packages that we will be using
import numpy as np                  # For arrays, matrices, and functions to operate on them
import matplotlib.pyplot as plt     # For showing plots
import pandas as pd                 # For data handling
from sklearn.datasets import load_digits

# Load the dataset
digits = load_digits()
```

```python
# Print the dataset
digits
```

```python
# Print the shape of the images
digits.data.shape
```

```python
# Print the shape of the target
digits.target.shape
```

```python
# Select one of the digits
i = 345
```

```python
# Show the data matrix of the selected digit
digits.images[i]
```

```python
# Show the number of the selected digit 
digits.target[i]
```

```python
# Display the digit
plt.figure(1, figsize=(3, 3))
plt.imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation="nearest")
plt.title('Number: ' + str(digits.target[i]) )
plt.show()
```

# Convert to a pandas dataframe


```python
dataset = pd.DataFrame(np.column_stack([digits['data'], digits['target']]), columns=digits['feature_names'] + ['target'])
```
